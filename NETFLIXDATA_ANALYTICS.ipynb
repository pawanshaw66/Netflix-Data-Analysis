#Q16. Create a bar chart showing the distribution of TV Shows and Movies by release year.


```python
grouped_data = df.groupby(['Year', 'Category']).size().unstack(fill_value=0)
```

```python
import matplotlib.pyplot as plt
grouped_data.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Distribution of TV Shows and Movies by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Category')
plt.tight_layout()
plt.show()
```

#Q17. Plot a pie chart illustrating the percentage of different content types (e.g., TV Shows, Movies).

```python
content = df['Category'].value_counts()
colors = ['violet', 'green']
plt.pie(content,labels=content.index, autopct='%1.1f%%', startangle=140,colors=colors)
plt.title('Percentage of Different Content Types')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
